## Introduction

An intriguing photograph of a pink Trabant car in front of a gray-green house wall, enigmatic lockers in the middle of a residential area or prefabricated buildings enthroned on a hill like a fortress: The images shown are part of the image collection of the Scientific Collections at the Leibniz Institute for Research on Society and Space (IRS). Among the hundreds of thousands of photos stored on the shelves of the archive there are numerous exciting motifs of which nobody knows exactly what they show. On which corner of the house did the Trabant stand? What purpose did the lockers once fulfill and how did they work? Above which place does the new housing estate tower? Who created the artwork on the façade of the tower block and when could the photo have been taken?

Questions like these do not only concern the employees of the Scientific collections, which are focussed on the Building and Planning History of the GDR. The shelves and boxes of many otherspecialized collections and small municipal archives contain nameless treasures that are just waiting to be made accessible by profound experts in the field so that they can be added to the fund of public knowledge. However, in view of the wealth of material, this work can rarely be carried out by the archivists themselves. Out of this need, the Scientific Collections developed the 'CitizenArchives' project together with the company Programmfabrik GmbH, developer of the digital asset management software 'FYLR', in order to jointly develop not only a product, but also a process that could be used to harness the commitment of volunteers. The project was a resounding success, with thousands of image descriptions being received via an online input interface within a short time. But how could we, the participatiBng archivists and scientists, ensure that our online interface was not hijacked by pranksters and scammers? What if politically motivated people tried to perpetuate a false image of the past in the archive by using faked image descriptions?

Spoiler alert: Our users did not try to trick us. But how did we know?

## The problem of document-rich, but resource-poor archives

At the Scientific Collections we often don't find answers to questions that are important for a detailed understanding and description of our pictures. We generally do know who took a photo and usually also in which city or in which of the former 15 GDR districts it was taken. However, we often do not have any further information on the photo collections that came to us via the predecessor Institute for Urban Design and Architecture (ISA) or the now almost 150 donors of materials - former architects and planners of the GDR. In many cases, the photographs in the form of prints, slides and negatives are unsorted and sparsely labeled or not labeled at all.

There is usually a lack of resources for a comprehensive indexing of digitized collections, for which extensive research into individual images is often necessary. This leaves it almost impossible to make them searchable and findable, meaning that historical treasures inevitably remain untapped and inaccessible for education, research and the public. Although users can scroll through our portal online, using visual cues to find images, indexing the motifs using metadata, i.e. adding titles, descriptions, keywords and names of people and places, makes them so much more accessible and valuable. The photographs, taken over four decades throughout East Germany and beyond, offer a unique insight not only into the history of architecture and urban development, but also into everyday life in the former GDR.

## State of research: Citizen science in historical research

After Citizen Science approaches initially were mainly used in the natural sciences, involving citizens in the process of analyzing  archive material is no longer a new and innovative phenomenon in Memory institutions  (<cite id="yo6n9"><a href="#zotero%7C3819060%2FWDFCRNG9">(Vohland et al., 2021)</a></cite>,  <cite id="eizol"><a href="#zotero%7C3819060%2FRUIN8NEB">(Tauginienė et al., 2020)</a></cite>). Many archives have experimented with this and made good experiences in the process. Usually the term ‘crowdsourcing’ is used to describe the contribution of knowledge to the description of objects in archives and scientific collections by citizens. (Huber et al. 2020; Graf 2019; Ridge 2016) 
The digitisation of inventories in particular has opened up new and promising opportunities. (Haumann und Paul 2022)


# Approach

A first attempt was made to create an input mask for citizen scientists in FYLR itself. However, we came up against the limits of the software, which is not designed for multiple users to enter different information about the same object. The programme is also not designed for collecting metadata in a separate 'protected area' or for low-threshold access without registration. The programme also lacks important functions for the citizen science aspect, such as a forum for direct exchange with the volunteers or areas for describing the project and its scientific background as well as for explanations on how to use it.


So in addition to the digitisation and preparation of initial test collections for the project and a needs and user analysis, it was necessary to find an alternative way of entering data in FYLR. When researching best practice examples of citizen science projects in the humanities and history, which are still in the minority compared to natural science projects, we repeatedly came across the citizen science platform 'Zooniverse'.


Zooniverse is probably the largest and best-known platform for citizen science projects. The web portal, hosted by the Citizen Science Alliance, headquartered at the renowned University of Oxford, allows anyone to create collaborative projects on a wide range of research topics within a given framework. Most of the existing projects deal with natural scientific topics. For example, volunteers can classify plants and animals or identify stars. However, there are also projects on historical issues that are dedicated to describing old postcards or transcribing information on architectural photographs. One of the best-known projects with a historical context in the Citizen Science landscape is also represented on Zooniverse: Every Name Counts from the Arolsen Archives, the world's largest archive on the victims and survivors of National Socialism. Here, users are asked to transcribe documents that shed light on the fates of concentration camp prisoners, forced laborers, people persecuted in other ways and so-called displaced persons.
